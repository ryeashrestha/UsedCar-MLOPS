```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = "used_cars.csv"
df = pd.read_csv(file_path)

# Display basic info and first few rows
df_info = df.info()
```

```python
# Missing values
missing = df.isnull().sum()
print("Missing Values:\n", missing)
```

```python

sns.histplot(df['price'], bins=50, kde=True)
plt.title("Distribution of Used Car Prices")
plt.xlabel("Price")
plt.show()
```

```python
# Histogram of model years
df['model_year'].hist(bins=20)
plt.title("Distribution of Model Years")
plt.xlabel("Year")
plt.ylabel("Frequency")
plt.show()
```

```python
df['brand'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Car Brands in Dataset")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(x='model_year', y='price', data=df)
plt.xticks(rotation=45)
plt.title("Price by Model Year")
plt.show()
```

```python
sns.boxplot(x='fuel_type', y='price', data=df)
plt.title("Price by Fuel Type")
plt.show()
```
